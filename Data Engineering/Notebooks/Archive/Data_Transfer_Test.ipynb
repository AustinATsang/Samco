```python
from pyspark.sql.functions import*
from pyspark.sql.types import StructType,StructField,StringType,ArrayType,IntegerType

edf = spark.sql("SELECT * FROM ecr2_test.ECR_List_Clean")
pdf = spark.sql("SELECT * FROM ecr2_test.ProdTable")

edf = edf.where(col('Stage_Mech') == 'WIP')
edf = edf.sort(edf['ECR#'].desc())
display(edf)
display(pdf)
```

```python
drawing_schema = ArrayType(
    StructType()
    .add(StructField('__id',StringType(),True))\
    .add(StructField('Product',StringType(),True))\
    .add(StructField('Drawing',StringType(),True))\
    .add(StructField('Current',StringType(),True))\
    .add(StructField('Ref',StringType(),True))\
    .add(StructField('New',StringType(),True))\
    .add(StructField('Prod',StringType(),True))\
    .add(StructField('Column1',StringType(),True))
)
```

```python
newdf = edf.select(
    col('ECR#'),\
    explode(from_json(col('Drawing_Mech'),drawing_schema)).alias('Drawing_Mech')
)
```

```python
newdf = newdf.withColumn('Product', col('Drawing_Mech.Product'))
newdf = newdf.withColumn('Drawing', col('Drawing_Mech.Drawing'))
newdf = newdf.withColumn('Current', col('Drawing_Mech.Current'))
newdf = newdf.withColumn('Ref', col('Drawing_Mech.Ref'))
newdf = newdf.withColumn('New', col('Drawing_Mech.New'))
newdf = newdf.withColumn('Production_Order', col('Drawing_Mech.Prod'))
newdf = newdf.withColumn('Production_Cost', col('Drawing_Mech.Column1'))
display(newdf)
```
